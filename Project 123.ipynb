```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output


cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
commuters_to_work = [3200000, 2100000, 1700000, 1500000, 1300000]
commuters_from_work = [2800000, 1800000, 1500000, 1300000, 1100000]


app = dash.Dash(__name__)


app.layout = html.Div([
    html.H1('Commuting Trends Dashboard'),
    html.Div([
        dcc.Graph(id='commuting-bar-chart')
    ]),
    html.Div([
        dcc.Dropdown(
            id='city-dropdown',
            options=[{'label': city, 'value': city} for city in cities],
            value='New York'
        )
    ]),
    html.Div(id='selected-city-stats')
])
```

```python
@app.callback(
    Output('selected-city-stats', 'children'),
    [Input('city-dropdown', 'value')]
)
def update_selected_city_stats(selected_city):
    city_index = cities.index(selected_city)
    to_work = commuters_to_work[city_index]
    from_work = commuters_from_work[city_index]
    return f"Commuters to work: {to_work}, Commuters from work: {from_work}"

```

```python
@app.callback(
    Output('commuting-bar-chart', 'figure'),
    [Input('city-dropdown', 'value')]
)
def update_bar_chart(selected_city):
    city_index = cities.index(selected_city)
    data = [
        {'x': ['To Work', 'From Work'], 'y': [commuters_to_work[city_index], commuters_from_work[city_index]], 'type': 'bar'}
    ]
    layout = {'title': f'Commuting Statistics for {selected_city}'}
    return {'data': data, 'layout': layout}
```
